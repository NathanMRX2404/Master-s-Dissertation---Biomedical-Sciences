```python
import pandas as pd
pd.set_option('display.max_columns', None)

# Read and print shape of uncalibrated data (linear)
uncal_results = "results/sage/FULL_OS_PXD001468.tsv"
df_uncal_results = pd.read_csv(uncal_results, sep='\t')
print("Lineair data uncalibrated:                     ", df_uncal_results.shape)

# Filtered uncalibrated data based on spectrum_q (linear)
df_uncal_results_q = df_uncal_results[df_uncal_results["spectrum_q"] < 0.01]
print("Lineair data uncalibrated q:                   ", df_uncal_results_q.shape)

LRcal_results = "results/sage/FULL_OS_PXD001468_LRcal.tsv"
df_LRcal_results = pd.read_csv(LRcal_results, sep='\t')
print("Lineair data LR calibrated:                    ", df_LRcal_results.shape)

# Filtered uncalibrated data based on spectrum_q
df_LRcal_results_q = df_LRcal_results[df_LRcal_results["spectrum_q"] < 0.01]
print("Lineair data LR calibrated q:                  ", df_LRcal_results_q.shape)

# Read and print shape of uncalibrated data
RFcal_results = "results/sage/FULL_OS_PXD001468_RFcal.tsv"
df_RFcal_results = pd.read_csv(RFcal_results, sep='\t')
print("Lineair data RF calibrated:                    ", df_RFcal_results.shape)

# Filtered uncalibrated data based on spectrum_q
df_RFcal_results_q = df_RFcal_results[df_RFcal_results["spectrum_q"] < 0.01]
print("Lineair data RF calibrated q:                  ", df_RFcal_results_q.shape)

print("")

# Read and print shape of uncalibrated data (non_linear)
uncal_results_nonlin = "results/sage/FULL_OS_PXD032235.tsv"
df_uncal_results_nonlin = pd.read_csv(uncal_results_nonlin, sep='\t')
print("Lineair data uncalibrated:                     ", df_uncal_results_nonlin.shape)

# Filtered uncalibrated data based on spectrum_q (non_linear)
df_uncal_results_nonlin_q = df_uncal_results_nonlin[df_uncal_results_nonlin["spectrum_q"] < 0.01]
print("Lineair data uncalibrated q:                   ", df_uncal_results_nonlin_q.shape)

LRcal_results_nonlin = "results/sage/FULL_OS_PXD032235_LRcal.tsv"
df_LRcal_results_nonlin = pd.read_csv(LRcal_results_nonlin, sep='\t')
print("Lineair data LR calibrated:                    ", df_LRcal_results_nonlin.shape)

df_LRcal_results_nonlin_q = df_LRcal_results_nonlin[df_LRcal_results_nonlin["spectrum_q"] < 0.01]
print("Lineair data LR calibrated q:                  ", df_LRcal_results_nonlin_q.shape)

RFcal_results_nonlin = "results/sage/FULL_OS_PXD032235_RFcal.tsv"
df_RFcal_results_nonlin = pd.read_csv(RFcal_results_nonlin, sep='\t')
print("Lineair data RF calibrated:                    ", df_RFcal_results_nonlin.shape)

df_RFcal_results_nonlin_q = df_RFcal_results_nonlin[df_RFcal_results_nonlin["spectrum_q"] < 0.01]
print("Lineair data RF calibrated q:                  ", df_RFcal_results_nonlin_q.shape)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

LowerLim = 125.5
UpperLim = 129.5
peak_value1 = 125.896648
peak_value2 = 128.094963
peak_value3 = 129.078979

# Assuming you have loaded your dataframes somewhere above
dataframes = [
    df_uncal_results_q[(df_uncal_results_q['mass error'] >= LowerLim -1) & (df_uncal_results_q['mass error'] <= UpperLim +1)], 
    df_LRcal_results_q[(df_LRcal_results_q['mass error'] >= LowerLim -1) & (df_LRcal_results_q['mass error'] <= UpperLim +1)], 
    df_RFcal_results_q[(df_RFcal_results_q['mass error'] >= LowerLim -1) & (df_RFcal_results_q['mass error'] <= UpperLim +1)]
]

titles = ['uncalibrated', 'LR calibrated', 'RF calibrated']

colors = [
    sns.color_palette()[0],
    sns.color_palette()[1],
    sns.color_palette()[2]
]

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Store y-values of annotated peaks
annotated_peak_values = []

# Plot histograms for each dataframe
for i, (df, title, color) in enumerate(zip(dataframes, titles, colors)):
    n, bins, patches = axs[i].hist(df['mass error'], bins=np.arange(LowerLim, UpperLim, 0.025), color=color, edgecolor='black')
    axs[i].set_ylim(0, 630)
    axs[i].set_xlim(LowerLim, UpperLim)


    # Find the bin where the peak_value falls
    bin_idx1 = np.where(bins <= peak_value1)[0][-1]
    bin_idx2 = np.where(bins <= peak_value2)[0][-1]
    bin_idx3 = np.where(bins <= peak_value3)[0][-1]
    
    annotated_peak_values.append(n[bin_idx1])  # Store y-value of annotated peak
    annotated_peak_values.append(n[bin_idx2])  # Store y-value of annotated peak
    annotated_peak_values.append(n[bin_idx3])  # Store y-value of annotated peak

    # Annotate the peak with 'deamidation' under the number
    axs[i].annotate(f'{peak_value1}\n iodination', xy=(bins[bin_idx1], n[bin_idx1]), xytext=(bins[bin_idx1] - 0.0150, n[bin_idx1] + 85),
                    arrowprops=dict(facecolor='black', arrowstyle='->'))
    if i == 1:
        axs[i].annotate(f'{peak_value2}\n addition of lysine\n due to transpeptidation ?', xy=(bins[bin_idx2], n[bin_idx2]), xytext=(bins[bin_idx2] - 2, n[bin_idx2] - 50),
                        arrowprops=dict(facecolor='black', arrowstyle='->', color = 'gray'), color = 'gray')
        axs[i].set_xlabel('mass shift (Da)')
        axs[i].set_title(f'PSMs around 127.5 Da - PXD001468')
    elif i == 2:
        axs[i].annotate(f'{peak_value2}\n addition of lysine\n due to transpeptidation ?', xy=(bins[bin_idx2], n[bin_idx2]), xytext=(bins[bin_idx2] - 2, n[bin_idx2] - 50),
                        arrowprops=dict(facecolor='black', arrowstyle='->', color = 'gray'), color = 'gray')
    else:
        axs[i].annotate(f'{peak_value2}\n addition of lysine\n due to transpeptidation ?', xy=(bins[bin_idx2], n[bin_idx2]), xytext=(bins[bin_idx2] - 1.5, n[bin_idx2] + 50),
                        arrowprops=dict(facecolor='black', arrowstyle='->', color = 'gray'), color = 'gray')
        axs[i].set_ylabel('frequency')

    axs[i].annotate(f'{peak_value3}\n acetylhypusine ?', xy=(bins[bin_idx3], n[bin_idx3]), xytext=(bins[bin_idx3] - 0.6, n[bin_idx3] + 85),
                    arrowprops=dict(facecolor='black', arrowstyle='->', color = 'gray'), color = 'gray')

    # Color the corresponding bins in red
    axs[i].patches[bin_idx1].set_facecolor('black')
    axs[i].patches[bin_idx2].set_facecolor('black')
    axs[i].patches[bin_idx3].set_facecolor('black')

# Add custom legend with adjusted position
legend_labels = [plt.Line2D([0], [0], color=color, lw=4) for color in colors]
fig.legend(legend_labels, titles, loc='upper center', bbox_to_anchor=(+0.605, +0.92))

# Print y-values of annotated peaks
print("Y-values of annotated peaks:", annotated_peak_values)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
